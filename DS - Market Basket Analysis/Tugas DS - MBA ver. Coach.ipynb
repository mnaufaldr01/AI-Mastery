```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

## Data Cleaning & Pre-Processing

```python
## Data Cleaning step

# replacing empty value with 0.
df.fillna(0,inplace=True)
```

```python
df
```

```python
# Data Pre-processing step

#for using aprori need to convert data in list format..
# transaction = [['apple','almonds'],['apple'],['banana','apple']]....
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
transactions[0]
```

## Market Basket Analysis

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df2, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
metrix = pd.DataFrame(rules)
metrix[['antecedents','consequents', 'support', 'confidence', 'lift']].sort_values('lift', ascending=False)
```

```python
metrix[['antecedents','consequents', 'support', 'confidence', 'lift']].sort_values('confidence', ascending=False)
```

## Kesimpulan
Dilihat dari analisis di atas, kita dapat melihat bahwa item yang paling besar kemungkinan akan dibeli bersamaan adalah *herb & pepper* dan *ground beef* dari nilai liftnya yang tinggi, namun dari nilai confidencenya kita juga mendapat insight bahwa kemungkinan orang untuk membeli kombinasi *herb & pepper* dan *ground beef* lebih tinggi jika orang tersebut membeli *herb & pepper*. Prinsip yang sama juga dapat kita simpulkan dari value support, confidence, dan lift dari kombinasi-kombinasi lainnya.
Kita juga menemukan bahwa *mineral water* merupakan konsekuensi yang paling populer dari kombinasi-kombinasi yang memiliki nilai confidence tertinggi merupakan kombinasi yang memiliki *mineral water* sebagai konsekuensi.
